```python
import os
from torchvision import transforms
from PIL import Image

# Paths
input_path = '/content/images/images/'  # Source folder
output_path = '/content/augmented_images_train'
os.makedirs(output_path, exist_ok=True)

# Debugging file detection
print("Files detected in the source folder:")
for root, _, files in os.walk(input_path):
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            print(file)

# Define basic augmentations for testing
augmentations = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(degrees=(-30, 30)),
    transforms.Resize((256, 256)),
    transforms.RandomCrop((224, 224)),
    transforms.ColorJitter(brightness=0.5),
    transforms.RandomGrayscale(p=0.2),
    transforms.ToTensor(),
])

image_count = 0
for root, _, files in os.walk(input_path):
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            img_path = os.path.join(root, file)
            try:
                # Open the image
                image = Image.open(img_path).convert("RGB")

                # Save original
                original_filename = f"{image_count}_original.jpg"
                image.save(os.path.join(output_path, original_filename))
                print(f"Saved: {original_filename}")
                image_count += 1

                # Save augmented
                for i in range(2):
                    augmented_image = augmentations(image)
                    augmented_filename = f"{image_count}_augmented_{i}.jpg"
                    augmented_image.save(os.path.join(output_path, augmented_filename))
                    print(f"Saved: {augmented_filename}")
                    image_count += 1
            except Exception as e:
                print(f"Error processing file {file}: {e}")

print(f"Augmented images saved to: {output_path}")
print(f"Number of images in destination folder: {len(os.listdir(output_path))}")
```

```python
import os
import torch
from torchvision import transforms, models
from PIL import Image
import numpy as np

input_path = '/content/augmented_images_train'
output_features_path = '/content/image_features.npy'

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
encoder = models.resnet18(pretrained=True)
encoder = torch.nn.Sequential(*list(encoder.children())[:-1])
encoder = encoder.to(device).eval()

preprocess = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

features = []
image_names = []

print("Extracting features...")
for root, _, files in os.walk(input_path):
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            img_path = os.path.join(root, file)
            try:
                image = Image.open(img_path).convert("RGB")
                input_tensor = preprocess(image).unsqueeze(0).to(device)

                with torch.no_grad():
                    feature = encoder(input_tensor).squeeze().cpu().numpy()

                features.append(feature)
                image_names.append(file)
                print(f"Processed: {file}")
            except Exception as e:
                print(f"Error processing file {file}: {e}")

features = np.array(features)
np.save(output_features_path, features)
print(f"Features saved to: {output_features_path}")

with open('/content/image_names.txt', 'w') as f:
    for name in image_names:
        f.write(name + '\n')

print("Feature extraction completed!")
```
